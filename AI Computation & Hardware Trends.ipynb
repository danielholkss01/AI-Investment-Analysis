```python
import pandas as pd

# Load the dataset
file_path = 'corporate-investment-in-artificial-intelligence-by-type.csv'  # Adjust the path as needed
data = pd.read_csv(file_path)

# Display the first few rows to confirm successful loading
data.head()
```

```python
# Check the data types and for missing values
data.info()

# Display basic statistics of the dataset
data.describe()
```

```python
# Drop the 'Code' column as it contains no useful data
data_cleaned = data.drop(columns=['Code'])

# Confirm the changes
data_cleaned.head()
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

# Filter the data for 'Merger/Acquisition'
merger_acquisition_data = data_cleaned[data_cleaned['Entity'] == 'Merger/Acquisition']

# Plot the investment trend over the years for 'Merger/Acquisition'
plt.figure(figsize=(12, 6))
plt.plot(merger_acquisition_data['Year'], merger_acquisition_data['Total corporate investment - inflation adjusted'], marker='o', label='Merger/Acquisition')
plt.title('Total Corporate Investment in AI: Merger/Acquisition Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Corporate Investment (Inflation Adjusted)')
plt.legend()
plt.grid(True)

# Save the plot as a PNG file
plt.savefig('merger_acquisition_investment_trend.png')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Load the dataset
file_path = 'ai-performance-knowledge-tests-vs-training-computation.csv'  # Adjust the path as needed
data_performance = pd.read_csv(file_path)

# Display the first few rows to confirm successful loading
data_performance.head()
```

```python
# Check the data types and for missing values
data_performance.info()

# Display basic statistics of the dataset
data_performance.describe()
```

```python
# If there are columns with many missing values or irrelevant data, drop them
# For example, if 'Code' column is irrelevant
data_performance_cleaned = data_performance.drop(columns=['Code'])

# Confirm the changes
data_performance_cleaned.head()
```

```python
import matplotlib.pyplot as plt

# Plot AI performance vs. training computation
plt.figure(figsize=(12, 6))
plt.scatter(data_performance_cleaned['Training computation (petaFLOP)'], data_performance_cleaned['MMLU avg'], color='blue', label='AI Models')

plt.title('AI Performance (MMLU avg) vs. Training Computation (petaFLOP)')
plt.xlabel('Training Computation (petaFLOP)')
plt.ylabel('AI Performance (MMLU avg)')
plt.legend()
plt.grid(True)

# Save the plot as a PNG file
plt.savefig('ai_performance_vs_training_computation.png')

# Show the plot
plt.show()
```

# AI Investment and Performance Analysis Report

## Introduction
This report provides an analysis of corporate investments in artificial intelligence (AI) over the years and examines the relationship between AI model performance and the computational resources used for training. The datasets used in this analysis are:
1. Corporate Investment in AI by Type
2. AI Performance on Knowledge Tests vs. Training Computation

## Data Analysis

### Corporate Investment in AI

**Trends in Mergers and Acquisitions:**
- A significant increase in investment from 2018 to 2021.
- The highest investment recorded in 2021, followed by a slight decrease in 2022.

**Investment Variability:**
- Variability in investment amounts year over year, indicating fluctuating market conditions and strategic priorities.

![Merger Acquisition Investment Trend](merger_acquisition_investment_trend.png)

### AI Performance vs. Training Computation

**Performance vs. Computation:**
- A positive correlation between training computation and AI performance.
- Higher training computation generally results in better performance on knowledge tests.

**Trends Over Time:**
- Increasing computational resources dedicated to training AI models over recent years.
- Performance improvements are more noticeable with larger computational investments.

![AI Performance vs Training Computation](ai_performance_vs_training_computation.png)
